**Step 4: Read the file**

```python
import pandas as pd
def load_housing_data(filename=filename):
   return pd.read_csv(filename)
```

**Step 5: Create a variable that contains all the text**

```python
newtext=''
for t in DataF['Comments']:
  newtext+=t

newtext
```

**Step 6: Configure the word cloud and remove Stopwords**

```python
from wordcloud import WordCloud, STOPWORDS
import sys, os
```

```python
wc=WordCloud(
    background_color='white',
    stopwords=STOPWORDS,
    height=600,
    width=400

)
```

```python
wc.generate(newtext)
```

**Step 7:Display the word cloud and save it in a file**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,12))
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show
```

```python
wc.to_file('Wcloud2.png')
```
